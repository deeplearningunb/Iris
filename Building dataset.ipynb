```python
import twint
import nest_asyncio
nest_asyncio.apply()
import pandas as pd
```

```python
def get_user_tweets(user):
    c = twint.Config()
    c.Username = user
    c.Retweets = False
    c.Custom['tweet'] = ['tweet']
#     c.Since = datetime.strftime(datetime.now() - timedelta(1), '%Y-%m-%d')
#     c.Until = datetime.strftime(datetime.now(), '%Y-%m-%d')
    c.Store_csv= True
    c.Store_object = True
    c.Format = ''
    c.Output = 'depressives.csv'
    twint.run.Search(c)
```

```python
def search_tweets(keyword):
    c = twint.Config()
    c.Search = keyword
    c.Custom['tweet'] = ['tweet']
#     c.Since = datetime.strftime(datetime.now() - timedelta(1), '%Y-%m-%d')
#     c.Until = datetime.strftime(datetime.now(), '%Y-%m-%d')
    c.Lang = 'pt'
    c.Store_csv= True
    c.Store_object = True
    c.Output = 'random_tweets.csv'
    c.Limit = 20
    twint.run.Search(c)
```

```python
expressions = ['"desanimo" "tristeza" -"excelente" -"globo"']
for exp in expressions:
    search_tweets(exp)
```

```python
# lendo datasets
depressives = pd.read_csv('depressives.csv', index_col=0)
random = pd.read_csv('random_tweets.csv', index_col=0)
```

```python
bad_words = ['pic.twitter', 'http']
```

```python
for word in bad_words:
    depressives = depressives[~depressives.tweet.str.contains(word)]
    random = random[~random.tweet.str.contains(word)]
```
